```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
composite_info = pd.read_csv('data/CT_composite.csv')
death_info = pd.read_csv('data/deaths_by_county.csv')
death_info
```

```python
# plot test -- to incorporate in loop
def unemployment_and_resident_deaths(county):
    df = composite_info[composite_info["Jurisdiction"] == county]
    plt.plot(df["Year"],df["ResidentDeaths"])
    plt.plot(df["Year"],df["%Unemployed"] * 10)
    plt.plot(df["Year"],df[""])
```

```python
unemployment_and_resident_deaths('Middlesex')
```

```python
def residentDeathsNorm():
    for jurisdiction in list(composite_info["Jurisdiction"].unique()):
        df = composite_info[composite_info["Jurisdiction"] == jurisdiction]
        plt.plot(df["Year"], df["ResidentDeathsNorm"] / df["unempNorm"], label=jurisdiction)
    plt.legend()
```

```python
residentDeathsNorm()
```

```python
def deathsInCountyNorm():
    for jurisdiction in list(composite_info["Jurisdiction"].unique()):
        df = composite_info[composite_info["Jurisdiction"] == jurisdiction]
        plt.plot(df["Year"], df["DeathsInCountyNorm"])
```

```python
deathsInCountyNorm()
```
